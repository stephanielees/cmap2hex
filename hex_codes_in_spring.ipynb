```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import matplotlib.patches as mpatch
import numpy as np
```

```python
CMAP_NAME = 'spring'
plt.get_cmap(CMAP_NAME)
```

```python
cmap = plt.get_cmap(CMAP_NAME)

hexes = [mcolors.to_hex(cmap(val)) for val in np.linspace(0, 1, 250)]
```

```python
_, axs = plt.subplots(50, 5, figsize=(10, 100), constrained_layout=True)
for i, ax in enumerate(axs.flat):
  ax.add_patch(mpatch.Rectangle((0, 0), 1, 1, color=hexes[i]))
  ax.set_title(hexes[i])
  ax.set_axis_off()
```
